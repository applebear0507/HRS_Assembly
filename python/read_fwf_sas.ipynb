```python
years = [2002, 2004, 2006, 2008, 2012, 2014, 2016]
def read_hrs_all_years(years, directory, variables_to_look_for):
    for i in range(len(years)):
        yr = years[i]
        print('reading files for survey year: ' + str(yr))
        if i == 0:
            year_frame = sas_read_year(directory, variables_to_look_for, yr)
        else:
            temp_df = sas_read_year(directory, variables_to_look_for, yr)
            year_frame = year_frame.append(temp_df)
        print('\n')
    return year_frame
    
output = read_hrs_all_years(years, directory, variables_to_look_for)
output
```
